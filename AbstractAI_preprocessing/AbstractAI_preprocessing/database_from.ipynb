```python
import pandas as pd

file_list = [
    "/content/drive/MyDrive/computer_sciences.xlsx", "/content/drive/MyDrive/material_sciences.xlsx", "/content/drive/MyDrive/engineering.xlsx"
]

y = ['Title', 'Abstract', 'Cited by']  ## common columns
MaxRowsPerFile = 1500

combined_df = pd.DataFrame()

for file_name in file_list:
    df = pd.read_excel(file_name, usecols=y)
    df = df.head(MaxRowsPerFile)
    combined_df = pd.concat([combined_df, df], ignore_index=True)


combined_df = combined_df.sort_values(by='Cited by', ascending=False)

combined_df.to_csv('/content/drive/MyDrive/combined_data_sorted.csv', index=False)
```
